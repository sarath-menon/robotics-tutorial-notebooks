```julia
using CairoMakie
using CSV
using DataFrames
```

```julia
tbl = (x = [1, 1, 1,2,2,2],
       height = 1:1:6,
       grp = [1, 2,  3, 1,2, 3],

       )

barplot(tbl.x, tbl.height,
        stack = tbl.grp,
        color = tbl.grp,
        axis = (xticks = (1:2, ["left", "middle"]),
                title = "Stacked bars"),
        )
```
