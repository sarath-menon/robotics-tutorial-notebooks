## Bode Plot

```julia
P = tf(1, [1, 0.2, 1])
C = pid(0.2, 1)

# loop transfer function
L = P*C

# stability margins in bode plot
marginplot(L)
```

## Nyquist plot

```julia
# plant
_0 = 1 # [rad/s]
 = 0.4

# controller parameters
k_p = 1

P = tf(1, [1, 2**_0, _0^2]) 
C = pid(k_p,0,0)
L = P*C

nyquistplot(L, label="without delay")
nyquistplot!(L * delay(0.5), label="with delay")
```
