# General systems

### First order system

```julia
a = 0.5
P = tf(1, [1, a])

# Simulate 20 seconds step response
res = step(P, 20) 
plot(res)
```

```julia
bodeplot(P)
```

```julia
si = stepinfo(res)
```

### Second order system

```julia
_0 = 0.8 # [rad/s]
 = 0.4
P = tf(1, [1, 2**_0, _0^2])

# Simulate 20 seconds step response
res = step(P, 20) 
plot(res)
```

```julia
bodeplot(P)
```

### Double integrator

```julia
P = tf(1, [1,0,0])

# Simulate 20 seconds step response
res = step(P, 20) 
plot(res)
```

# Delayed systems

### first order system with delay

```julia
a = 0.5
P = tf(1, [1, a])
P_delayed = tf(1, [1, a])* delay(2)

# Simulate 20 seconds step response
res = step(P_delayed, 20) 
plot(res)
```

```julia
bodeplot(P, label="without delay")
bodeplot!(P_delayed, label="with delay")
```

### second order system with delay

```julia
_0 = 0.8 # [rad/s]
 = 0.4
P = tf(1, [1, 2**_0, _0^2])
P_delayed = P * delay(2)

# Simulate 20 seconds step response
res = step(P_delayed, 20) 
plot(res)
```

```julia
bodeplot(P, label="without delay")
bodeplot!(P_delayed, label="with delay")
```

# Filters

### Lowpass filter

```julia
_0 = 2 # [rad/s]
 = 0.8

P = tf(_0^2, [1, 2**_0, _0^2])
bodeplot(P)
```

### Bandpass filter

```julia
_0 = 2 # [rad/s]
 = 0.8

P = tf([2**_0, 0], [1, 2**_0, _0^2])
bodeplot(P)
```

### Highpass filter

```julia
_0 = 2 # [rad/s]
 = 0.8

P = tf([1, 0, 0], [1, 2**_0, _0^2])
bodeplot(P)
```

# Examples

### Electric motor

```julia
# motor parameters
k_I = 1
J = 2
c = 1

# controller parameters
k_p = 1

P = tf(k_I, [J, c, 0]) 
C = pid(k_p,0,0)

# loop T.F
L = P*C;
```

```julia
nyquistplot(L)
```

### Linearized cart pendulum

```julia
P = tf(1, [1, 0, -1]) 

# controller parameters
k_p = 2
k_d = 0.5

C = pid(k_p,0,k_d)
L = P*C

isstable(feedback(L))
```

```julia
nyquistplot(L, unit_circle=true)
```

```julia
response = impulse(feedback(L), 20)
plot(response)
```
