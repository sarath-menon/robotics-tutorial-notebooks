```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
```

## Facility layout

```python
# Data generation
n = 4
x = np.random.rand(n)
y = np.random.rand(n)

x_var = cp.Parameter(len(x), value=x)
y_var = cp.Parameter(len(y), value=y)

# Plotting
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random (x, y) points')
plt.show()
```

### Minimize avg dist to center

```python

x_c = cp.Variable(1)
y_c = cp.Variable(1)

cost = cp.norm(x_var - x_c, p=2) + cp.norm(y_var - y_c, p=2)
prob = cp.Problem(cp.Minimize(cost))
prob.solve()

plt.scatter(x_c.value, y_c.value)
plt.scatter(x, y, color='r', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Optimized (x, y) points')
plt.show()
```

### Minimize max dist to center

```python
x_c = cp.Variable(1)
y_c = cp.Variable(1)

cost = cp.norm(x_var - x_c, p='inf') + cp.norm(y_var - y_c, p='inf')
prob = cp.Problem(cp.Minimize(cost))
prob.solve()

plt.scatter(x_c.value, y_c.value)
plt.scatter(x, y, color='r', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Optimized (x, y) points')
plt.show()
```
