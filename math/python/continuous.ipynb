```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use('dark_background')
```

# Uniform distribution

## Cumulative distribution

```python
x = np.linspace(-5,10,1500)
loc = -3; scale = 6
f = stats.uniform.pdf(x,loc,scale) 
F = stats.uniform.cdf(x,loc,scale)

fig, axs = plt.subplots(2, gridspec_kw={'hspace': 0.5})
fig.suptitle('Uniform Distribution')
axs[0].plot(x, f)
axs[0].set_title('Probability Density Function')
axs[1].plot(x, F)
axs[1].set_title('Cumulative Distribution Function')
plt.show()
```

## Distribution statistics

```python
 # to generate 1000 uniform random numbers
 a = 0; b = 1;
 X = stats.uniform.rvs(a,b,size=1000)
 plt.hist(X);
```

```python
mean = np.mean(X)
var = np.var(X)
Med = np.median(X)
Mod = stats.mode(X)
print(f"Mean: {mean:.2f}, Variance: {var:.2f}, Median: {Med:.2f}, Mode: {Mod}")
```

## Gaussian

```python
x  = np.linspace(-10,10,1000)
mu = 0; sigma = 1;
f  = stats.norm.pdf(x,mu,sigma)

plt.plot(x,f)
```

## Mixture of gaussians

```python
x = np.linspace(-10, 10, 1000)
mixtures = 3

# mixture weights and normalize
pi = np.random.rand(mixtures) 
pi = pi/np.sum(pi)

mu = np.random.uniform(-5, 5, size=len(pi)) # means
sigma = np.random.uniform(0.1, 1.5, size=len(pi)) # standard deviations

# Calculate the PDF for each Gaussian and sum them
f = np.zeros_like(x)
for i in range(len(pi)):
    f += pi[i] * stats.norm.pdf(x, mu[i], sigma[i])

plt.plot(x, f)
plt.show()
```
