```python
from collections import Counter
```

## Load data

```python
with open('tinyshakespere.txt', 'r') as file:
    contents = file.read()

# Create a list of unique characters in the contents
unique_chars = list(set(contents))

char_frequency = dict(sorted(Counter(contents).items()))

# Normalized frequency
total_chars = sum(char_frequency.values())
normalized_char_frequency = {char: freq/total_chars for char, freq in char_frequency.items()}
```

## Historgrm

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 4))
plt.bar(char_frequency.keys(), char_frequency.values())
plt.xlabel('Characters')
plt.ylabel('Normalized Frequency')
plt.title('Character Frequency Histogram')
plt.show()
```
