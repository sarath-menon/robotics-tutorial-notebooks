```julia
using OrdinaryDiffEq, GLMakie
using Distributions
using DiffEqBase.EnsembleAnalysis
using  Measurements

GLMakie.activate!(inline=true)
```

# Determinstic

```julia
function lotka_volterra(du,u,p,t)
  du[1] = p[1] * u[1] - p[2] * u[1]*u[2]
  du[2] = -p[3] * u[2] + p[4] * u[1]*u[2]
end
```

```julia
θ = [1.5,1.0,3.0,1.0]
u0 = [1.0;1.0]
tspan = (0.0,10.0)

prob1 = ODEProblem(lotka_volterra,u0,tspan,θ)
sol = solve(prob1,Tsit5(), abstol=1e-8, reltol=1e-8);
y = hcat(sol.u...)

fig = lines(sol.t, y[1,:])
lines!(sol.t, y[2,:])

display(fig)
```

# Probabilistic

```julia
# create distribution 
θ = (Normal(0.5,0.1), Normal(5,1), Normal(3,0.3), Normal(3,0.4))

# draw samples from distrib
_θ = rand.(θ)

prob1 = ODEProblem(lotka_volterra,u0,tspan,_θ)
sol = solve(prob1,Tsit5(), abstol=1e-8, reltol=1e-8)
y = hcat(sol.u...)

fig = lines(sol.t, y[1,:])
lines!(sol.t, y[2,:])
display(fig)
```

## Ensemble of solutions

```julia
prob_func = function (prob,i,repeat)
  remake(prob,p=rand.(θ))
end

ensemble_prob = EnsembleProblem(ODEProblem(lotka_volterra,u0,tspan,θ),
                                prob_func = prob_func)

sol = solve(ensemble_prob,Tsit5(),EnsembleThreads(),trajectories=1000);

ensemble_sol = EnsembleSummary(sol)

# compute statistics
q_low_ensemble = hcat(ensemble_sol.qlow.u...)
q_high_ensemble = hcat(ensemble_sol.qhigh.u...)
q_mean_ensemble = hcat(ensemble_sol.u.u...)
```

```julia
id = 1

fig = Figure()
Axis(fig[1, 1])

lines!(ensemble_sol.t, q_mean_ensemble[id,:], color=:black) 
band!(ensemble_sol.t, q_low_ensemble[id,:], q_high_ensemble[id,:], color=RGBAf(100,0,0,0.1))

display(fig)
```

# Uncertainity

```julia
gaccel = 9.79 ± 0.02; # Gravitational constants
L = 1.00 ± 0.01; # Length of the pendulum

#Initial Conditions
u₀ = [0 ± 0, π / 3 ± 0.02] # Initial speed and initial angle
```

```julia
tspan = (0.0, 6.3)

#Define the problem
function simplependulum(du,u,p,t)
    θ  = u[1]
    dθ = u[2]
    du[1] = dθ
    du[2] = -(gaccel/L) * sin(θ)
end

#Pass to solvers
prob = ODEProblem(simplependulum, u₀, tspan)
sol = solve(prob, Tsit5(), reltol = 1e-6)
```

```julia
# Determinstic part 
# waste.val
# waste.err


fig = Figure()
Axis(fig[1, 1])

y = hcat(sol.u...)

id = 2

val_vec = Measurements.value.(y[id,:])
error_vec = Measurements.uncertainty.(y[id,:])
    
xs = sol.t
ys = val_vec

errorbars!(xs, ys, error_vec,
    color = :red,
    whiskerwidth = 10)

lines!(xs,ys)

display(fig)
    
```
