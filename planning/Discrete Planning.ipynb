```julia

function plot_grid(length, width)

    n_cells = length * width

    fig = Figure()
    rects = fig[1:width, 1:length] = [Box(fig, color = :red) for i in 1:n_cells]

    # set start cell color to green
    rects[1,1].color = :green

    display(fig)
end
```

```julia
let

length = 10
width = 10
    
plot_grid(length, width)

end
```
